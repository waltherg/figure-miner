# Figure Miner Prototype

Idea put forward by Ross Mounce: 

> See if we can extract data from figures (e.g. the coordinates of an x,y plot) and provide that data in a machine-readable form.

> see [https://github.com/hack4ac/hack4ac.github.io/wiki/Teams-&-Ideas#figure-mining--enrichment](https://github.com/hack4ac/hack4ac.github.io/wiki/Teams-&-Ideas#figure-mining--enrichment)

This is prototype code that fetches a PLoS image from fighshare and detects contours in the image.

The result at the bottom shows that most elements of the plot are detected -- further processing of the obtained contours should enable us to extract the axes, their ticks and labels, and (hopefully all) data points.


## Setup

```python
import numpy as np
import cv2
import cv
from matplotlib import pylab as pl
```

## Download and Load Image

Following [this tutorial](http://www.neuroforge.co.uk/index.php/getting-started-with-python-a-opencv) and [this IPython Notebook](http://nbviewer.ipython.org/urls/gist.github.com/schwehr/4694343/raw/98159da50f6a5bb0b4b846751843581bfabcf19d/opencv-learning.ipynb) for basic image downloading and loading.

```python
!wget http://files.figshare.com/1094381/Figure_1.tif
```

```python
image = cv2.imread('Figure_1.tif')
```

```python
pl.imshow(image)
```

## Convert Image to Grayscale

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
pl.imshow(image_gray)
```

## Detect and Display Contours

Tutorials for contour detection:

* [http://opencvpython.blogspot.co.uk/2012/06/hi-this-article-is-tutorial-which-try.html](http://opencvpython.blogspot.co.uk/2012/06/hi-this-article-is-tutorial-which-try.html)

The example below should be modified to detect disks or ellipses. May find inspiration here:

* [http://stackoverflow.com/questions/14058366/opencv-dominoes-circular-spots-disks-detection](http://stackoverflow.com/questions/14058366/opencv-dominoes-circular-spots-disks-detection)

And links therein in:

* [http://stackoverflow.com/questions/4785419/detection-of-coins-and-fit-ellipses-on-an-image](http://stackoverflow.com/questions/4785419/detection-of-coins-and-fit-ellipses-on-an-image)

* [http://opencv.willowgarage.com/documentation/cpp/imgproc_structural_analysis_and_shape_descriptors.html#cv-fitellipse](http://opencv.willowgarage.com/documentation/cpp/imgproc_structural_analysis_and_shape_descriptors.html#cv-fitellipse)

* [http://www.aishack.in/2010/07/tracking-colored-objects-in-opencv/](http://www.aishack.in/2010/07/tracking-colored-objects-in-opencv/)

```python
contours, hierarchy = cv2.findContours(image_gray,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
# TODO: figure out how to clone images
image = cv2.imread('Figure_1.tif')

for index in range(len(contours)):
    cv2.drawContours(image=image, contours=contours, contourIdx=index, color=(255, 0, 0))
```

```python
pl.imshow(image, interpolation='nearest', aspect='auto')
```
